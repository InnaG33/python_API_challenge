```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
# If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
import json

city_weather = []
print("Beginning Data Retrieval")
print("--"*30)
for index in range(len(cities)):
 #   try:
    print(f"Processing Record {index} | {cities[index]}")
    query_url = f"http://api.openweathermap.org/data/2.5/weather?q={cities[index]}&appid={weather_api_key}"
    post_response = requests.get(query_url)
    city_weather.append(post_response.json())
    print("--"*10)
 #   except (KeyError, IndexError):
 #       print("City not found. Skipping...")
        
print("--"*30)
```

```python
print(json.dumps(city_weather, indent=4, sort_keys=True))
```

```python
weather_data = pd.DataFrame()
weather_data ["Country"] = ""
weather_data ["City name"] = ""
weather_data ["Date"] = ""
weather_data ["Lat"] = ""
weather_data ["Lng"] = ""
weather_data ["Clouds"] = ""
weather_data ["Mean_T"] = ""
weather_data ["Feels_like_T"] = ""
weather_data ["Max_T"] = ""
weather_data ["Min_T"] = ""
weather_data ["Wind speed"] = ""
weather_data ["Humidity"] = ""
weather_data ["Pressure"] = ""
weather_data
```

```python
for index in range(len(cities)):
    try:
        weather_data.loc[index, "Country"] = city_weather[index]['sys']['country']
        weather_data.loc[index, "City name"] = city_weather[index]['name']
        weather_data.loc[index, "Date"] = city_weather[index]['dt']
        weather_data.loc[index, "Lat"] = city_weather[index]["coord"]["lat"]
        weather_data.loc[index, "Lng"] = city_weather[index]["coord"]["lon"]
        weather_data.loc[index, "Mean_T"] = city_weather[index]["main"]["temp"]
        weather_data.loc[index, "Feels_like_T"] = city_weather[index]["main"]["feels_like"]
        weather_data.loc[index, "Max_T"] = city_weather[index]["main"]["temp_max"]
        weather_data.loc[index, "Min_T"] = city_weather[index]["main"]["temp_min"]
        weather_data.loc[index, "Clouds"] = city_weather[index]["clouds"]["all"]
        weather_data.loc[index, "Wind speed"] = city_weather[index]["wind"]["speed"]
        weather_data.loc[index, "Humidity"] = city_weather[index]["main"]["humidity"]
        weather_data.loc[index, "Pressure"] = city_weather[index]["main"]["pressure"]
    except (KeyError, IndexError):
        weather_data = weather_data.append(pd.Series(0, index=weather_data.columns), ignore_index=True)

weather_data
```

```python
weather_data = weather_data.loc[~(weather_data==0).all(axis=1)]
weather_data = weather_data.reset_index(drop=True)
weather_data
```

```python
# Function that plots [x,y] scatter data with a liner regression line

def plot_response_linrreg(x, y, x_axis, y_axis, title):
    regress_values = []
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    for i in x:
        regress_values.append(i * slope + intercept)
    
    plt.figure(figsize=(7,5))
    plt.scatter(x, y, s=40, facecolor="skyblue", color='darkblue')  
    plt.plot(x, regress_values, "r-", linewidth=1)
    
    plt.title(title, fontsize=16)
    plt.xlabel(x_axis, fontsize=14)
    plt.ylabel(y_axis, fontsize=14)
    plt.annotate(line_eq, xy=((min(x)+abs(min(x)*0.1)), (min(y)+abs(min(y)*0.1))), fontsize=14, color="red")
    plt.grid()
    plt.show()
```

```python
# Test plotting function
import random 
x=[]
y=[]
for i in range (10):
    x.append(random.uniform(-100, 100))
    y.append(random.uniform(-100, 100))

plot_response_linrreg(x, y, "x_axis", "y_axis", "Title")
```
